```python
import numpy as np
import matplotlib.pyplot as plt



def distance(x, y):
    return np.sqrt(x**2+y**2)


def montecarlo(N):
    ans = []
    x = np.random.uniform(0, 1, N)
    y = np.random.uniform(0, 1, N)

    inCircle = 0
    inSquare = 0

    coords = list(zip(x, y))
    x_values = [pair[0] for pair in coords]
    y_values = [pair[1] for pair in coords]

    for i in range(len(coords)):
        d = distance(x_values[i],y_values[i])
        if d <= 1:
            color = "red"
            inCircle+=1
            if N==10**4:
                plt.scatter(x_values[i], y_values[i], c=color)
        else:
            color = "blue"
            inSquare+=1
            if N == 10**4:
                plt.scatter(x_values[i], y_values[i], c=color)

    pie = 4*(inCircle/N)
    error = abs((np.pi - pie)/np.pi)*100
    ans.append(pie)
    ans.append(error)
    if N ==10**4:
        plt.show()
    return ans

l1= montecarlo(10**3)
l3=montecarlo(10**5)
l4=montecarlo(10**6)
l2 = montecarlo(10**4)

print("n =  10^3", ": pi =", "{: .6f}".format(
    l1[0]), "& error =", "{: .4f}".format(l1[1]), "%")
print("n =  10^4", ": pi =", "{: .6f}".format(
    l2[0]), "& error =", "{: .4f}".format(l2[1]), "%")
print("n =  10^5", ": pi =", "{: .6f}".format(
    l3[0]), "& error =", "{: .4f}".format(l3[1]), "%")
print("n =  10^6", ": pi =", "{: .6f}".format(
    l4[0]), "& error =", "{: .4f}".format(l4[1]), "%")
 

```
